2. Hitung mean, median, std, min, max
3. Ubah ke matriks 4x5
4. Tampilkan shape dan transpose

## Penjelasan Bagian NumPy

Pada bagian ini dibuat 20 nilai acak menggunakan fungsi np.random.rand(20). Secara default, fungsi tersebut menghasilkan angka dalam rentang 0 sampai 1. Agar nilainya lebih variatif dan menyerupai skala penilaian umum, setiap angka kemudian dikalikan 100 sehingga berada pada kisaran 0 hingga 100.

Nilai-nilai yang dihasilkan selanjutnya ditampilkan dengan pembulatan dua angka desimal menggunakan round(2). Pembulatan ini bertujuan agar output terlihat lebih rapi dan mudah dibaca.

Setelah data tersedia, dilakukan perhitungan beberapa statistik dasar untuk menganalisis karakteristiknya. Statistik yang dihitung meliputi rata-rata (mean), median (nilai tengah), standar deviasi (sebagai ukuran penyebaran data), nilai minimum, dan nilai maksimum. Perhitungan ini membantu memberikan gambaran umum mengenai distribusi angka yang telah dibuat.

Kemudian, data yang semula berbentuk array satu dimensi diubah menjadi matriks berukuran 4 baris dan 5 kolom menggunakan metode reshape(4,5). Transformasi ini menunjukkan bagaimana data numerik dapat disusun dalam bentuk tabel dua dimensi.

Terakhir, ditampilkan ukuran matriks menggunakan atribut shape untuk mengetahui dimensinya. Selain itu, digunakan operasi transpose (.T) untuk menukar posisi baris dan kolom. Proses ini penting dalam pengolahan data, khususnya pada machine learning, karena struktur matriks sering kali perlu disesuaikan dengan kebutuhan analisis atau algoritma tertentu.

```python
arr = np.random.rand(20) * 100

print("Array (20 angka):")
print(arr.round(2))

mean_val = arr.mean()
median_val = np.median(arr)
std_val = arr.std()
min_val = arr.min()
max_val = arr.max()

print("\nStatistik:")
print("Mean  :", round(mean_val, 2))
print("Median:", round(median_val, 2))
print("Std   :", round(std_val, 2))
print("Min   :", round(min_val, 2))
print("Max   :", round(max_val, 2))

mat = arr.reshape(4, 5)

print("\nMatriks 4x5:")
print(mat.round(2))
print("\nShape matriks:", mat.shape)
print("\nTranspose:")
print(mat.T.round(2))
```

## B. Pandas (25%)

Tugas:
1. Buat DataFrame 10 mahasiswa (Nama, Nilai)
2. Tambahkan kolom Status: Lulus jika nilai ≥ 75, selain itu Tidak Lulus
3. Tampilkan:
   - 3 nilai tertinggi
   - Rata-rata nilai kelas
   - Jumlah mahasiswa lulus

## Penjelasan Bagian Pandas ( Data Frame Mahasiswa)
Pada bagian ini dibuat daftar nama mahasiswa secara otomatis menggunakan perulangan dengan range(10). Dengan cara ini, sistem menghasilkan nama berformat Mahasiswa_1 hingga Mahasiswa_10 tanpa perlu mengetikkannya satu per satu.

Selanjutnya, dibuat data nilai secara acak menggunakan fungsi np.random.randint(50, 101, size=10). Perintah tersebut menghasilkan 10 angka acak dengan rentang nilai mulai dari 50 sampai 100.

Nama dan nilai yang telah dibuat kemudian digabungkan ke dalam sebuah DataFrame menggunakan pd.DataFrame(). DataFrame berperan sebagai tabel data yang memiliki dua kolom utama, yaitu “Nama” dan “Nilai”, sehingga data lebih terstruktur dan mudah dianalisis.

Setelah itu ditambahkan kolom baru bernama “Status”. Kolom ini dibuat menggunakan fungsi np.where() untuk menentukan keterangan kelulusan berdasarkan nilai yang diperoleh. Jika nilai mahasiswa lebih dari atau sama dengan 75, maka diberikan status “Lulus”. Sebaliknya, jika nilainya di bawah 75, maka statusnya adalah “Tidak Lulus”.

Terakhir, DataFrame ditampilkan untuk melihat keseluruhan tabel yang berisi nama mahasiswa, nilai yang diperoleh, serta status kelulusannya.

```python
nama = [f"Mahasiswa_{i+1}" for i in range(10)]
nilai = np.random.randint(50, 101, size=10)  # 50–100

df_mhs = pd.DataFrame({
    "Nama": nama,
    "Nilai": nilai
})

df_mhs["Status"] = np.where(df_mhs["Nilai"] >= 75, "Lulus", "Tidak Lulus")

print("DataFrame Mahasiswa:")
df_mhs
```

## Penjelasan Bagian Pandas ( Mencari Nilai Tertinggi Mahasiswa)

Perintah df_mhs.nlargest(3, "Nilai") digunakan untuk memilih tiga mahasiswa dengan nilai paling tinggi. Fungsi nlargest() akan mengurutkan data berdasarkan kolom yang ditentukan secara menurun (dari terbesar ke terkecil), kemudian mengambil sejumlah data sesuai yang diminta, yaitu tiga baris teratas.

Selanjutnya, display(top3) dipakai untuk menampilkan hasil tersebut dalam bentuk tabel yang lebih terstruktur dan mudah dibaca dibandingkan tampilan print biasa.

Perintah df_mhs["Nilai"].mean() berfungsi menghitung nilai rata-rata seluruh mahasiswa dalam kelas. Nilai rata-rata ini memberikan gambaran umum mengenai performa akademik kelas secara keseluruhan.

Kemudian, round(rata2, 2) digunakan untuk membulatkan hasil rata-rata hingga dua angka di belakang koma. Tujuannya agar angka yang ditampilkan terlihat lebih rapi dan informatif.

Bagian (df_mhs["Status"] == "Lulus").sum() digunakan untuk menghitung jumlah mahasiswa yang berstatus lulus. Prosesnya dimulai dengan mengecek setiap baris pada kolom “Status”, menghasilkan nilai True atau False, lalu menjumlahkan seluruh nilai True (yang merepresentasikan mahasiswa lulus).

Terakhir, int(jml_lulus) memastikan bahwa hasil yang ditampilkan berupa bilangan bulat tanpa angka desimal, sehingga tampil lebih jelas dan sesuai dengan konteks jumlah orang.

```python
top3 = df_mhs.nlargest(3, "Nilai")
print("Top 3 Nilai:")
display(top3)

rata2 = df_mhs["Nilai"].mean()
print("\nRata-rata nilai kelas:", round(rata2, 2))

jml_lulus = (df_mhs["Status"] == "Lulus").sum()
print("Jumlah mahasiswa lulus:", int(jml_lulus))
```

## Visualisasi (Bar Histogram)
plt.figure(figsize=(6,4))
Digunakan untuk menentukan ukuran tampilan grafik agar proporsional dan nyaman dilihat, tidak terlalu besar maupun terlalu kecil.

counts, bins = np.histogram(df_mhs["Nilai"], bins=5)
Pada bagian ini, fungsi np.histogram() dipakai untuk menghitung jumlah data dalam setiap kelompok rentang nilai. Parameter bins=5 berarti nilai mahasiswa dibagi menjadi lima interval. Hasilnya berupa:

counts → jumlah data (frekuensi) pada tiap rentang
bins → batas nilai dari masing-masing rentang

bin_centers = 0.5 * (bins[1:] + bins[:-1])
Langkah ini menghitung titik tengah dari setiap rentang nilai. Titik tengah tersebut digunakan sebagai posisi pada sumbu horizontal agar grafik garis bisa merepresentasikan distribusi data dengan lebih akurat.

plt.plot(bin_centers, counts, marker='o')
Data frekuensi kemudian ditampilkan dalam bentuk grafik garis. Titik-titik pada grafik diberi tanda lingkaran (marker='o') agar setiap nilai frekuensi terlihat jelas.

plt.title("Grafik Distribusi Nilai")
Memberikan judul pada grafik untuk menjelaskan bahwa visualisasi ini menunjukkan sebaran nilai mahasiswa.

plt.xlabel("Rentang Nilai")
Memberi keterangan pada sumbu horizontal yang merepresentasikan kelompok rentang nilai.

plt.ylabel("Frekuensi")
Memberikan label pada sumbu vertikal yang menunjukkan jumlah mahasiswa dalam setiap rentang nilai.

plt.grid(True, linestyle="--", alpha=0.5)
Menambahkan garis bantu agar pembacaan grafik menjadi lebih mudah, dengan tampilan garis putus-putus yang tidak terlalu mencolok.

plt.tight_layout()
Menyesuaikan tata letak elemen grafik secara otomatis agar tidak saling bertabrakan.

plt.savefig("pertemuan01_grafik_distribusi.png")
Menyimpan hasil grafik ke dalam file gambar berformat PNG.

plt.show()
Menampilkan grafik pada layar output.

print("✅ Grafik distribusi disimpan: pertemuan01_grafik_distribusi.png")
Menampilkan pesan konfirmasi bahwa grafik telah berhasil dibuat dan disimpan.


```python
import numpy as np

plt.figure(figsize=(6,4))

# Hitung histogram
counts, bins = np.histogram(df_mhs["Nilai"], bins=5)

# Ambil titik tengah tiap bin
bin_centers = 0.5 * (bins[1:] + bins[:-1])

# Plot sebagai grafik garis
plt.plot(bin_centers, counts, marker='o')

plt.title("Grafik Distribusi Nilai")
plt.xlabel("Rentang Nilai")
plt.ylabel("Frekuensi")

plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()

plt.savefig("pertemuan01_grafik_distribusi.png")
plt.show()

print("✅ Grafik distribusi disimpan: pertemuan01_grafik_distribusi.png")
```
